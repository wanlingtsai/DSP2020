# Exercise
Make a square signal at 1100 Hz and make a wave that samples it at 10000 frames per second. If you plot the spectrum, you can see that most of the harmonics are aliased. When you listen to the wave, can you hear the aliased harmonics?發出1100 Hz的方波，並產生以每秒10000幀的速度對其採樣的波形。如果繪製頻譜圖，則可以看到大多數諧波都是混疊的。當您聽波時，您能聽到混疊諧波嗎？


# Solution
Here's the square wave:
這是方波：


```python
square = thinkdsp.SquareSignal(1100).make_wave(duration=0.5, framerate=10000)
```


Here's what the spectrum looks like:

```python
square.make_spectrum().plot()
```

The fundamental and the first harmonic are in the right place, but the second harmonic, which should be 5500 Hz, is aliased to 4500 Hz. The third, which should be 7700 Hz, is aliased to 2300 Hz, and so on.

When you listen to the wave, you can hear these aliased harmonics as a low tone.

基波和一次諧波在正確的位置，但是二次諧波（應為5500 Hz）被別名為4500 Hz。第三個應該是7700 Hz，別名是2300 Hz，依此類推。


```python
square.make_audio()

```

The tone I hear most prominently is at 300 Hz. If you listen to this 300 Hz sine wave, you might hear what I mean.我最明顯聽到的聲音是300 Hz。如果您收聽此300 Hz正弦波，您可能會聽到我的意思。


```python
thinkdsp.SinSignal(300).make_wave(duration=0.5, framerate=10000).make_audio()
```
